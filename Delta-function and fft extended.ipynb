```python
# Подробно изучим поведение fft для "дельта-функции". Поскольку наш интеграл на неё очень похож, можно увидеть похожую картину.
```

```python
import numpy as np
from numpy import fft
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
def delta_demo(dx, length, offset_position):
    a = np.zeros(2**length)
    print(2**length, 'точек')
    a[len(a)//2 - offset_position] = 1
    # значения функции, напротив номеров точек
    plt.plot(a, 'g')
    plt.ylabel('func value')
    plt.xlabel('sample num', color='r')
    plt.show()
    
    # значения функции, напротив значений x
    x_space = np.linspace(0, dx * 2**length,  num = 2**length, endpoint = False)
    plt.plot(x_space, a, 'g')
    plt.ylabel('func value')
    plt.xlabel('sample coordinates n
               ', color='r')
    plt.show()
    
    # значения преобразованной, напротив частот в герцах
    xi_space = fft.fftfreq(2**length, d = dx)
    
    b = fft.fft(a)

    plt.plot(xi_space, b.real, '-')
    plt.show()
    
    plt.plot(xi_space, b.imag, 'r-')
    plt.show()
widgets.interact(delta_demo,
                 dx = widgets.FloatSlider(min= 0.01,max= 0.1,step=0.01,value=0.01),
                 length = widgets.IntSlider(min= 2,max= 10,step=1,value=3),
                 offset_position=widgets.IntSlider(min= - 100,max= 100,step=1,value=0))
```
